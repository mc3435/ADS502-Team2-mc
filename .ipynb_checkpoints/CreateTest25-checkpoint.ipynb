```python
#Convert to numeric function, drop years and geo variables
def getNumericCol(df):
    # Excluding date column from the dataset just to have numerical data in order to find out the outliers
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    df = df.select_dtypes(include=numerics )
    
    # Remove any year(yr) columns for numeric ananlysis
    df = df[df.columns.drop(list(df.filter(regex='yr')))]
    
    # Remove any year(yr) columns for numeric ananlysis
    df = df[df.columns.drop('id')]
    
    return df
```
